```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium import webdriver
import os                       
import requests
import numpy as np
import pandas as pd
import time



def GetCollageAndUnivercityNamesAndAddress(UnivercityFileName,CollageFileName):
    dfUniversity = pd.DataFrame(columns=['Name','Address'])
    dfCollege = pd.DataFrame(columns=['Name','Address'])
    dictionaryAcademicType = {'University': '', 'College': ''}
    link_Universities=r"https://che.org.il/המוסדות-להשכלה-גבוהה/#section-79442"
    link_College1=r"https://che.org.il/המוסדות-להשכלה-גבוהה/#section-79446"
    link_College2=r"https://che.org.il/המוסדות-להשכלה-גבוהה/#section-79448"
    link_College3=r"https://che.org.il/המוסדות-להשכלה-גבוהה/#section-79452"
    links = [link_Universities, link_College1, link_College2, link_College3]

   
    try:
        for item in links:      
            driver.get(item)       
            driver.implicitly_wait(20)
            sectionName=item.split("#")
            link = driver.find_element(by=By.ID, value=sectionName[1])
            title=link.find_element_by_css_selector("div.title").text
            subLink=link.find_element_by_css_selector("div.content")
            subsublink=subLink.find_elements_by_css_selector("a")
            print(title)
            linkes = list()
            if str(title) == "אוניברסיטאות":
                for itemLink in subsublink:
                    linkes.append(itemLink.get_attribute("href"))
                dictionaryAcademicType['University'] = linkes
            else:
                for itemLink in subsublink:
                    linkes.append(itemLink.get_attribute("href"))
                dictionaryAcademicType['College'] = linkes

        for itemAcademicType in dictionaryAcademicType:
            typeName = str(itemAcademicType)
            linkStringList = dictionaryAcademicType[itemAcademicType]
            for itemLink in linkStringList:
                try:
                    driver.get(itemLink)
                    driver.implicitly_wait(20)
                    name = driver.find_elements_by_class_name("yellow-title")
                    address = driver.find_elements_by_class_name("col-xs-9")
                   

                    if typeName == "University":
                        dfUniversity = dfUniversity.append({'Name':  name[0].text, 'Address': address[0].text}, ignore_index=True)
                        print(dfUniversity)
                    else:
                        dfCollege = dfCollege.append({'Name': name[0].text, 'Address': address[0].text},ignore_index=True)
                        print(dfCollege)
                    time.sleep(0.5)
                except:
                    print("Invalid URL")
    except:
        print("Invalid URL")
    driver.quit()  
    dfUniversity.to_csv(UnivercityFileName, mode='a', encoding='utf-8-sig', index=False)
    dfCollege.to_csv(CollageFileName, mode='a', encoding='utf-8-sig', index=False)
    return dfUniversity , dfCollege
```
